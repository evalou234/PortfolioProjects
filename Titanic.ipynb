```python
df1 = df.copy() #dataframe copy

df1.isnull().sum()
```

```python
df1['age'] = df1['age'].fillna(np.mean(df1['age']))
df1.isnull().sum()
#all null values of Age column replace with the mean of the column
```

```python
# replace NAN Values(float)
df1['pclass'].fillna(df1['pclass'].mode()[0], inplace=True)
df1.isnull().sum()
df1['survived'].fillna(df1['survived'].mode()[0], inplace=True)
df1.isnull().sum()
df1['sibsp'].fillna(df1['sibsp'].mode()[0], inplace=True)
df1.isnull().sum()
df1['parch'].fillna(df1['parch'].mode()[0], inplace=True)
df1.isnull().sum()
df1['ticket'].fillna(df1['ticket'].mode()[0], inplace=True)
df1.isnull().sum()
df1['fare'].fillna(df1['fare'].mode()[0], inplace=True)
df1.isnull().sum()
df1['boat'].fillna(df1['boat'].mode()[0], inplace=True)
df1.isnull().sum()
df1['body'].fillna(df1['body'].mode()[0], inplace=True)
df1.isnull().sum()
```

```python
#replace NA Values(object)
df1['name'].value_counts()
df1['name'].mode()
df1['name'].fillna(df1['name'].mode()[0], inplace=True)

df1['sex'].value_counts()
df1['sex'].mode()
df1['sex'].fillna(df1['sex'].mode()[0], inplace=True)


df1['embarked'].value_counts()
df1['embarked'].mode()
df1['embarked'].fillna(df1['embarked'].mode()[0], inplace=True)


df1['cabin'].value_counts()
df1['cabin'].mode()
df1['cabin'].fillna(df1['cabin'].mode()[0], inplace=True)

df1['home.dest'].value_counts()
df1['home.dest'].mode()
df1['home.dest'].fillna(df1['home.dest'].mode()[0], inplace=True)
df1.isnull().sum()
```
